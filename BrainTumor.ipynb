```python
# Initialize lists
dataset = []
label = []
INPUT_SIZE = 244

labels_dict = {
    'notumor': 0,
    'glioma': 1,
    'meningioma': 2,
    'pituitary': 3
}

# Loop through each subfolder and process images
for subfolder, label_value in labels_dict.items():
    subfolder_path = os.path.join(image_directory, subfolder)
    image_names = os.listdir(subfolder_path)

    for image_name in image_names:
        if image_name.split('.')[-1].lower() == 'jpg':  # Ensure it's a JPG file
            # Load and process the image
            image = cv2.imread(os.path.join(subfolder_path, image_name), cv2.IMREAD_GRAYSCALE)  
            image = Image.fromarray(image)
            image = image.resize((INPUT_SIZE, INPUT_SIZE))

            # Append the processed image and label to the lists
            dataset.append(np.array(image))
            label.append(label_value)

# Convert the lists to numpy arrays if needed for further processing
dataset = np.array(dataset)
label = np.array(label)

print("Shape of dataset:", dataset.shape)
```

Data Augmentation


```python
# Data augmentation configuration
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python

# Define the list of subfolders
categories = ['notumor', 'glioma', 'meningioma', 'pituitary']

# Loop through each subfolder (category) and augment images
for subfolder in categories:
    subfolder_path = os.path.join(image_directory, subfolder)
    os.makedirs(subfolder_path, exist_ok=True)  # Ensure the directory exists

    # Create a flag file to indicate augmentation is done
    flag_file = os.path.join(subfolder_path, 'augmentation_done.txt')
    if os.path.exists(flag_file):
        continue  # Skip if augmentation is already done

    # List all images in the subfolder
    for image_name in os.listdir(subfolder_path):
        if image_name.lower().endswith(('.jpg', '.jpeg', '.png')):  # Ensure it's an image file
            # Load and process the image
            img_path = os.path.join(subfolder_path, image_name)
            img = load_img(img_path)  # Load image
            x = img_to_array(img)  # Convert to numpy array
            x = x.reshape((1,) + x.shape)  # Reshape for data augmentation

            # Generate and save augmented images
            i = 0
            for batch in datagen.flow(x, batch_size=1, save_to_dir=subfolder_path, 
                                      save_prefix='aug', save_format='jpg'):
                i += 1
                if i > 2:  
                    break

    # Create the flag file to indicate augmentation is done
    with open(flag_file, 'w') as f:
        f.write('Augmentation done')

print("Augmentation completed.")
```

Splitting Data 

```python
# Define the paths
image_directory = './MRI/'
output_directory = './Split_Data/'

# Create training and testing directories
train_dir = os.path.join(output_directory, 'training')
test_dir = os.path.join(output_directory, 'testing')

```

```python
# Create the directory structure
for dir_type in [train_dir, test_dir]:
    for category in categories:
        os.makedirs(os.path.join(dir_type, category), exist_ok=True)

# Split and move files
for category in categories:
    category_path = os.path.join(image_directory, category)
    images = [f for f in os.listdir(category_path) if f.endswith('.jpg')]
    
    # Split into training and testing
    train_images, test_images = train_test_split(images, test_size=0.2, random_state=42)
    
    # Move training images
    for image in train_images:
        src = os.path.join(category_path, image)
        dst = os.path.join(train_dir, category, image)
        shutil.copyfile(src, dst)
    
    # Move testing images
    for image in test_images:
        src = os.path.join(category_path, image)
        dst = os.path.join(test_dir, category, image)
        shutil.copyfile(src, dst)

print("Data splitting and organizing complete!")
```

```python
import os

# Define the paths for training and testing directories
base_dir = './Split_Data/'  # Adjust this path if necessary

# Define the categories
categories = ['notumor', 'glioma', 'meningioma', 'pituitary']

# Initialize dictionaries to store counts
training_counts = {}
testing_counts = {}

# Loop through categories to count images in training and testing sets
for category in categories:
    # Define paths for training and testing folders for each category
    train_path = os.path.join(base_dir, 'training', category)
    test_path = os.path.join(base_dir, 'testing', category)
    
    # Count the number of images in training and testing folders
    training_image_count = len([f for f in os.listdir(train_path) if f.endswith('.jpg')])
    testing_image_count = len([f for f in os.listdir(test_path) if f.endswith('.jpg')])
    
    # Store counts in dictionaries
    training_counts[category] = training_image_count
    testing_counts[category] = testing_image_count

# Print the counts
print("Number of images in each training folder:")
for category, count in training_counts.items():
    print(f"{category}: {count}")

print("\nNumber of images in each testing folder:")
for category, count in testing_counts.items():
    print(f"{category}: {count}")
```

```python
# Function to get image shape
def get_image_shape(image_path):
    with Image.open(image_path) as img:
        return img.size  # Returns (width, height)

# Function to print shapes and counts
def print_image_shapes_and_counts(directory, categories):
    # Initialize a variable to count the total number of images
    total_images = 0
    
    for subfolder in categories:
        subfolder_path = os.path.join(directory, subfolder)
        
        # List all images in the subfolder
        image_files = [f for f in os.listdir(subfolder_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
        
        # Count the number of images
        num_images = len(image_files)
        total_images += num_images
        
        # Get the shape of a sample image
        if image_files:
            sample_image_path = os.path.join(subfolder_path, image_files[0])
            sample_image_shape = get_image_shape(sample_image_path)
        else:
            sample_image_shape = (0, 0)  # No images available
        
        print(f"\nCategory: {subfolder}")
        print(f"Number of images: {num_images}")
        print(f"Shape of images: {sample_image_shape}")
    
    # Print total number of images across all subfolders
    print(f"\nTotal number of images in {directory}: {total_images}")

# Print shapes and counts for training and testing directories
print("Printing shapes of images in training directory...")
print_image_shapes_and_counts(train_dir, categories)

print("Printing shapes of images in testing directory...")
print_image_shapes_and_counts(test_dir, categories)
```

```python
from PIL import UnidentifiedImageError
# Load a fixed number of images from each category
images = []
labels = []

for category in categories:
    category_path = os.path.join(train_dir, category)
    for img_name in os.listdir(category_path)[:4]:  # Load only the first 4 images from each category
        img_path = os.path.join(category_path, img_name)
        try:
            img = load_img(img_path, target_size=(244, 244))  # Adjust target_size if needed
            img_array = img_to_array(img) / 255.0  # Normalize to [0, 1]
            images.append(img_array)
            labels.append(category)
        except UnidentifiedImageError:
            print(f"Skipping file {img_name} as it is not a valid image.")

# Plot the images
plt.figure(figsize=(20, 20))
for i in range(len(images)):
    plt.subplot(4, 4, i + 1)
    plt.imshow(images[i])
    plt.title(labels[i], color='black', fontsize=16, fontweight='bold')
    plt.axis('off')
plt.show()
```
